```python
import kenlm
lang_model = kenlm.Model('language_model_dat/3-gram-words.arpa')
```

```python
folder_name = "language_model_dat/1-billion-word/training"
f_names = os.listdir(folder_name)
text_names = [os.path.join(folder_name, f_name) for f_name in f_names if not re.match(r"(.*)\.npy", f_name)]
```

```python
f = open("language_model_dat/processed-text-clean.txt", "a+")
for (i,fname) in enumerate(text_names):
    t_f = open(fname, "r")
    file_lines = t_f.readlines()
    for line in file_lines:
        f.write(clean_sentence(line))
    
    print "Done with text {}".format(i)
    print clean_sentence(file_lines[-1])

f.close()
```
